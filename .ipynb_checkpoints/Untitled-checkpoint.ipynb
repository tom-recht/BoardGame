```python
import matplotlib.pyplot as plt

# Parameters
num_episodes = 1000
initial_epsilon = 1.0
epsilon_decay = 0.995
min_epsilon = 0.1
epsilon_boost = 0.1  # Boost value

def plot(epsilon_boost):
    # List to store epsilon values
    epsilon_values = []

    # Compute epsilon values over episodes
    epsilon = initial_epsilon
    for episode in range(num_episodes):
        if episode % 100 == 0:
            epsilon = min(1.0, epsilon + epsilon_boost)
        else:
            epsilon = max(min_epsilon, epsilon * epsilon_decay)
        epsilon_values.append(epsilon)

    # Plotting epsilon values over episodes
    plt.figure(figsize=(10, 6))
    plt.plot(epsilon_values, label=f'Epsilon with Boost = {epsilon_boost}')
    plt.xlabel('Episode')
    plt.ylabel('Epsilon')
    plt.title('Epsilon Values Over Episodes')
    plt.legend()
    plt.grid(True)
    plt.show()
```
